```python
def cleanData(df):
    #convert all columns to object (ML string problems)
    #df.dtypes[df.dtypes.map(lambda x: x == 'object')]
    #Column with missing values
    missing_col = df.columns[df.isnull().any()].tolist()
    print(missing_col)
    df[missing_col] = df[missing_col].fillna((df[missing_col].mean()), inplace=True)
    #for item in missing_col:
        
    #    df4 = df[df[item].notnull()].groupby(item)
    #    df[[item]] = df[[item]].fillna(df4.median)
    #Transforme catheogorical data to numerical data
    col_to_conv = ['playerShort', 'club', 'leagueCountry', 'position', 'photoID', 'Alpha_3']
    for col in col_to_conv:
        items_cat = pd.Categorical(df[col], ordered= False) # default order: alphabetical
        df[col] = items_cat.labels # insert in datafram 
        #items = sorted(df[col].unique())
        #items_map = dict(zip(items, range(0, len(items) + 1)))
        #df[col] = df[col].map(items_map).astype(int)
    return df
        
```

```python
from itertools import chain, combinations
def powerset(iterable):
    "powerset([1,2,3]) --> () (1,) (2,) (3,) (1,2) (1,3) (2,3) (1,2,3)"
    s = list(iterable)
    return chain.from_iterable(combinations(s, r) for r in range(len(s)+1))
```

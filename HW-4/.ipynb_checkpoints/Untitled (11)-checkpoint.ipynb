```python
def cleanData(df):
    #convert all columns to object (ML string problems)
    df.dtypes[df.dtypes.map(lambda x: x == 'object')]
    #Column with missing values
    missing_col = df.columns[df.isnull().any()].tolist()
    for item in missing_col:
        df[[item]].fillna(df.groupby(item).transform('median'))
    #Transforme catheogorical data to numerical data
    col_to_conv = ['playerShort', 'club', 'leagueCountry', 'position']
    for col in col_to_conv:
        items_cat = pd.Categorical(df[col], ordered= False) # default order: alphabetical
        df[col] = items_cat.labels # insert in datafram 
        #items = sorted(df[col].unique())
        #items_map = dict(zip(items, range(0, len(items) + 1)))
        #df[col] = df[col].map(items_map).astype(int)
    return df
```

```python
df2 = cleanData(df1)
```

Since the data in here is in form of dyads, we would need to group by players.

Now let's aggregate the values per player for useful columns and drop the rest:

Before apllying ML techniques, we should normalize the data (the features for which it makes sense to normalize):

```python
def normalize(x):
    return (x - x.mean()) / x.std()
```

```python
cols_to_norm = ['height', 'weight', 'games', 'victories', 'defeats', 'ties', 'goals', 'redCards', 'yellowReds', 'yellowCards'] 
for col in cols_to_norm:
    aggr_df[col] = normalize(aggr_df[col])
```

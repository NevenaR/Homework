```python
# Drop the non worthing columns
df1.drop(['player', 'birthday', 'photoID', 'Alpha_3' ], inplace=True, axis=1)
```

```python
def cleanData(df):
    #convert all columns to object (ML string problems)
    df.dtypes[df.dtypes.map(lambda x: x == 'object')]
    #Column with missing values
    missing_col = df.columns[df.isnull().any()].tolist()
    for item in missing_col:
        df[[item]].fillna(df.groupby(item).transform('median'))
    #Transforme catheogorical data to numerical data
    col_to_conv = ['playerShort', 'club', 'leagueCountry', 'position']
    for col in col_to_conv:
        items_cat = pd.Categorical(df[col], ordered= False) # default order: alphabetical
        df[col] = items_cat.labels # insert in datafram 
        #items = sorted(df[col].unique())
        #items_map = dict(zip(items, range(0, len(items) + 1)))
        #df[col] = df[col].map(items_map).astype(int)
    return df
        
```

```python
df3 = cleanData(df1)
```

```python
df3.tail()
```

```python
plt.rc('figure', figsize=(25, 5))
# Set up the grid of plots
fig = plt.figure(figsize=fizsize_with_subplots)
fig_dims = (3, 2)

#Plot heights
plt.subplot2grid(fig_dims, (0, 0))
df3['height'].value_counts(sort=False).plot(kind='bar', 
                                         title='Heights distribution')

#Plot weights
plt.subplot2grid(fig_dims, (0, 1))
df3['weight'].value_counts(sort = False).plot(kind='bar', 
                                         title='Weights distribution')

#Plot positions
plt.subplot2grid(fig_dims, (1, 0))
df3['position'].value_counts(sort = False).plot(kind='bar', 
                                    title='Position distribution')

# Plot yellowCards counts
plt.subplot2grid(fig_dims, (1, 1))
df3['yellowCards'].value_counts(sort = False).plot(kind='bar', 
                                         title='yellowCards distribution')

# Plot yellowRedscounts
plt.subplot2grid(fig_dims, (2, 0))
df3['yellowReds'].value_counts(sort = False).plot(kind='bar', 
                                         title='yellowReds distribution')

# Plot redCards counts
plt.subplot2grid(fig_dims, (2, 0))
df3['redCards'].value_counts(sort = False).plot(kind='bar', 
                                       title='redCards distribution')

plt.tight_layout()
# problem with sorting labels, to be fixed ...
```

```python
def normalize(x):
    #return (x - x.mean()) / x.std()
    return (x - x.mean()) / (x.max() - x.min())
```

```python
# Here we normalize both height and weight
cols_to_norm = ['height', 'weight']
df3[['height_norm','weight_norm']] = df3[cols_to_norm].apply(lambda x: normalize(x)) 
```

```python
df3['IATEXPmean'] = df3[['meanIAT', 'meanExp']].mean(axis=1)
df3['IATEXPstd'] = df3['IATEXPmean'].std()
```
